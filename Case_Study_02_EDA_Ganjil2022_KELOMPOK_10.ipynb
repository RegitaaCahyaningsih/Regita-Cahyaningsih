1. Apakah ada dosen yang cenderung memberikan nilai cenderung lebih tinggi (atau lebih rendah) dibandingkan dosen lainnya?


```python
nilai_dosen = nilai.groupby(nilai['NIDN']).mean()[['nilai']]
nilai_dosen = nilai_dosen.rename({'nilai': 'mean_nilai'}, axis=1) 
nilai_dosen = pd.merge(dosen, nilai_dosen, left_on='NIDN', right_index=True)
nilai_dosen = nilai_dosen[['NIDN', 'nama', 'mean_nilai']]
sb = nilai.groupby(nilai['NIDN']).std()[['nilai']]
sb = sb.rename({'nilai': 'std_nilai'}, axis=1) 
nilai_dosen = nilai_dosen.set_index(['NIDN']) 
nilai_dosen = pd.concat([nilai_dosen, sb], axis=1)
nilai_dosen
```

**Interpretasi pada nomer 1:**

Bu Annisa memberikan nilai cenderung lebih rendah karena memiliki rata-rata nilai dan standar deviasi lebih kecil dibandingkan dengan dosen lainnya, tetapi tidak memiliki cukup bukti, karena hal ini belum dibuktikan pada pengajaran jumlah kelas dan hanya dugaan sementara.

Sedangkan, Pak Andri memberikan nilai cenderung lebih tinggi karena memiliki rata-rata nilai dan standar deviasi lebih besar dibandingkan dengan dosen lainnya, tetapi tidak memiliki cukup bukti, karena hal ini belum dibuktikan pada pengajaran jumlah kelas dan hanya dugaan sementara.

2. Menurut data yang ada, apakah terdapat perbedaan dalam pemberian nilai antara dosen dengan Pendidikan Terakhir S3 dan S2?


```python
pd_dosen = dosen[['NIDN','nama','Pendidikan_Terakhir']]
nilai_dosen2 = nilai[['NIDN', 'nilai']]
nilai_dosen2 = nilai.groupby(nilai['NIDN']).mean()[['nilai']]
nilai_dosen2 = nilai_dosen2.rename({'nilai': 'mean'}, axis =1)
std = nilai.groupby(nilai['NIDN']).std()[['nilai']]
std = std.rename({'nilai': 'standar deviasi'}, axis =1)
pd_dosen = pd_dosen.set_index(['NIDN'])
nilai_dosen2 = pd.concat([nilai_dosen2, std], axis=1)
pd_dosen = pd.concat([nilai_dosen2, pd_dosen], axis = 1)
pd_dosen
```

**Interpretasi pada nomor 2**:

Berdasarkan data terdapat perbedaan dalam pemberian nilai antara dosen dengan pendidikan terakhir S3 dan S2, dosen dengan pendidikan terakhir S2 cenderung memberikan nilai lebih rendah sedangkan dosen dengan pendidikan terakhir S3 memberikan nilai lebih tinggi.

Beradasarkan tabel diatas, kami dapat menyimpulkan bahwa diantara dosen dengan pendidikan terakhir S2, Bu Annisa termasuk dosen yang memberikan nilai terendah dengan mean senilai 60 dan standar deviasi 16. 3. Apakah terdapat perbedaan performa akademik antara mahasiswa yang berasal dari Madrasah/SMK dan SMU?


```python
asal_mahasiswa = mahasiswa[['nim', 'Asal_Sekolah']]
nilai2 = nilai[['nim','nilai']]
asal_mahasiswa = asal_mahasiswa.set_index(['nim'])
nilai2 = nilai2.groupby(nilai2['nim']).mean()
mahasiswa3 = pd.concat([asal_mahasiswa, nilai2], axis=1)
mahasiswa3 = mahasiswa3.reset_index(drop=False, inplace=False)
mahasiswa3 = mahasiswa3[["Asal_Sekolah",'nilai']]
mahasiswa3 = mahasiswa3.groupby(mahasiswa3['Asal_Sekolah']).mean()
mahasiswa3 = mahasiswa3.rename({'nilai':'rata rata nilai berdasarkan asal sekolah'}, axis=1)
mahasiswa3
```

**Interpretasi pada nomer 3:**

Berdasarkan data diatas, kami mengetahui bahwa SMK memiliki nilai rata - rata lebih tinggi dari pada MA dan SMU. Sehingga kami bisa menyimpulkan jika jenjang pendidikan tingkat SMK lebih terlatih karena banyak praktikum sehingga performa akademiknya lebih unggul daripada MA dan SMU.

4. Apakah ada perbedaan yang nampak di data terhadap kemampuan akademik antara mahasiswa dan mahasiswi?

5. Apakah ada perbedaan prestasi akademik yang nampak di data antar angkatan mahasiswa?


```python
angkatan_mahasiswa = mahasiswa[['nim', 'Angkatan']]
nilai2 = nilai[['nim', 'nilai']]
angkatan_mahasiswa = angkatan_mahasiswa.set_index(['nim'])
nilai2 = nilai2.groupby(nilai2['nim']).mean()
maba = pd.concat([angkatan_mahasiswa, nilai2], axis=1)
maba = maba.reset_index(drop=False, inplace=False)
maba = maba[["Angkatan",'nilai']]
maba = maba.groupby(maba['Angkatan']).mean()
maba = maba.rename({'nilai':'rata rata nilai berdasarkan Angkatan'}, axis=1)
maba
```
